```
import json
from tokenizers import Tokenizer

tok = Tokenizer.from_file("tokenizer.json")
eos = tok.token_to_id("</s>")  # adjust for your tokenizer

trunc_no_eos = 0
total = 0
for line in open("train.jsonl"):
    ex = json.loads(line)
    if ex.get("role") == "assistant":
        ids = tok.encode(ex["content"]).ids
        total += 1
        if ids[-1] != eos:
            trunc_no_eos += 1
print("assistant turns:", total, "no_eos:", trunc_no_eos)
```
